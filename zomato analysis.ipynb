```python
coun = pd.read_excel("Country-Code.xlsx")
coun.head()
```

```python
final_df = pd.merge(zom,coun,on='Country Code', how = 'left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df['Country'].value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
## pie plot for the top 3 countries using zomato
plt.pie(country_val[:3], labels=country_names[:3], autopct= "%1.2f%%")
plt.title("Top 3 countries using zomato")
plt.savefig('top_3.jpg')
```

Observation: Zomato maximum records or transactions are from India After that USA and then United Kingdom.

```python
final_df.columns
```

```python
Ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
Ratings
```

## Observations
1. When Rating is between 4.5 to 4.9 ----> Excellent
2. When Rating are between 4.0 to 4.4--->  very good
3. when Rating is between 3.5 to 3.9 ----> good
4. when Rating is between 3.0 to 3.4 ----> average
5. when Rating is between 2.5 to 2.9 ----> average
6. when Rating is between 2.0 to 2 4 ----> Poor

```python
plt.figure(figsize=[12,6])
sns.barplot(x="Aggregate rating", y = "Rating Count", hue = "Aggregate rating", data = Ratings)
plt.savefig("Ag_rc.jpg")
```

```python
Ratings['Rating color'].unique()
```

```python
plt.figure(figsize=[12,6])
sns.barplot(x="Aggregate rating", y = "Rating Count", hue = "Rating color", data = Ratings, palette=['blue', 'Red', 'Orange', 'Yellow', 'Green', 'Green'])
plt.savefig('Rc_RC.jpg')
```

# Observation:
1. Highest number of people has given 0 rating.
2. The average rating is in between 2.5-3.4.

```python
## count plot
sns.countplot(x = 'Rating color', data = Ratings, palette=['blue', 'Red', 'Orange', 'Yellow', 'Green', 'Green'])
plt.savefig('RC_cp.jpg')
```
